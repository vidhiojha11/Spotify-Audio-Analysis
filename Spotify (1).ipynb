```python
import pandas as pd
df= pd.read_csv(r"C:\Users\Vidhi\Downloads\Spotify_dataset.csv")
df.columns
```

```python
top_5_artists = (df.groupby('artists', as_index=False)['popularity'].mean().sort_values(by='popularity', ascending=False).head(5))
```

```python
top_5_artists = (
    df.groupby('artists')['popularity']
      .mean()
      .sort_values(ascending=False)
      .head(5)
)

print(top_5_artists)
```

```python
top_5_tracks = (
    df[['track_name', 'popularity']]
      .sort_values(by='popularity', ascending=False)
      .head(5)
      .set_index('track_name')  
)

print(top_5_tracks)
```

```python
import matplotlib.pyplot as plt
top_5_artists.plot.barh()
plt.show()
```

```python
top_5_tracks.plot.barh()
plt.show()
```

### DANCEABILITY

```python
top_5_artist_danceable_songs = (df[['danceability','artists','track_name']].sort_values(by='danceability', ascending = False).head(5))
```

```python
print(top_5_artist_danceable_songs)
```

```python
import seaborn as sns
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.figure(figsize = (12,7))
sns.barplot(x='danceability', y='artists', data = top_5_artist_danceable_songs, palette = colors)
plt.title("Artist with most danceable song")
plt.show()          
```

```python
top_5_tracks.plot.barh(
    colormap='viridis',
    figsize=(10,5)
)

plt.title("Top 5 Tracks by Popularity")
plt.xlabel("Popularity")
plt.ylabel("Track")
plt.show()
```

```python
top_5_tracks.plot.barh(
    color=['red', 'blue', 'green', 'orange', 'purple'],
    figsize=(10,5)
)

plt.title("Top 5 Tracks by Popularity")
plt.xlabel("Popularity")
plt.ylabel("Track")
plt.show()
```
